```python
#filter data for LA vs GSW

Bron_LA_GSW_mask = df_bron_games_all2d['MATCHUP'].str.lower().isin(['lal @ gsw', 'lal vs. gsw']) # creates a boolean pandas series based on the condition

type(Bron_LA_GSW_mask), df_bron_games_all2d[Bron_LA_GSW_mask].shape #only 11 games with GSW?
```

```python
df_bron_games_all2d[df_bron_games_all2d['MATCHUP'].str.contains('LAL @ GSW|LAL vs. GSW')].shape #trying it one more way to confirm
```

```python
#check for CLE games with GSW

Bron_CLE_GSW_mask = df_bron_games_all2d['MATCHUP'].str.lower().isin(['cle @ gsw', 'cle vs. gsw']) # creates a boolean pandas series based on the condition
type(Bron_CLE_GSW_mask), df_bron_games_all2d[Bron_CLE_GSW_mask].shape #only 11 games with GSW?
```

```python
#Looking at the range of plus minuses for Bron's CLE vs GSW
Bron_CLE_GSW_games = df_bron_games_all2d[Bron_CLE_GSW_mask]
Bron_CLE_GSW_games['PLUS_MINUS'].min(), Bron_CLE_GSW_games['PLUS_MINUS'].max()
```

```python
#looking at the number of times there were the plus minuses for the cleveland gsw games
counts = Bron_CLE_GSW_games['PLUS_MINUS'].value_counts().reset_index()
counts.columns = ['Value', 'Count']#titling the columns
counts = counts.sort_values(by='Value', ascending=False)
counts
```

```python
#binning the results

bins = range(Bron_CLE_GSW_games['PLUS_MINUS'].min() - 6, Bron_CLE_GSW_games['PLUS_MINUS'].max() + 6, 10)
binned_values = pd.cut(Bron_CLE_GSW_games['PLUS_MINUS'], bins)
binned_counts = binned_values.value_counts().reset_index()
binned_counts.columns = ['Bin', 'Count']

binned_counts = binned_counts.sort_values(by = 'Bin', ascending=False)
binned_counts
```
